```python
import pickle
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
params = {'beta1':1, 'beta0':-1, 's2_ep':1}
mcmc_120 = pickle.load(open('../model_mc_simul/MC_SIMUL_SLR_120.pkl','rb'))
mcmc_300 = pickle.load(open('../model_mc_simul/MC_SIMUL_SLR_300.pkl','rb'))
mcmc_500 = pickle.load(open('../model_mc_simul/MC_SIMUL_SLR_500.pkl','rb'))
mcmc_1000 = pickle.load(open('../model_mc_simul/MC_SIMUL_SLR_1000.pkl','rb'))
```

```python
mcmc_dict = {120:mcmc_120, 300:mcmc_300, 500:mcmc_500, 1000:mcmc_1000}
n = [120, 300, 500, 1000]

for j in n:
    mcmc = mcmc_dict[j]

    a = 0
    fig, ax = plt.subplots(1,3,figsize=(15,5))
    for i in range(3):
        key = list(mcmc.keys())[a]
        sns.distplot(mcmc[key], ax=ax[i], norm_hist=True, bins=30)
        kdeline = ax[i].lines[0]
        xs = kdeline.get_xdata()
        ys = kdeline.get_ydata()
        ax[i].vlines(params[key] ,0, ys.max())
        ax[i].set_title(f'Posterior distribution for {key}')

        a += 1

    plt.savefig(f'/Users/shinbo/Desktop/metting/LDA/meeting materials/20.12.24/density_{j}.png')
    plt.show()
```

```python
for j in n:
    mcmc = mcmc_dict[j]
    for param in ['beta0', 'beta1', 's2_ep']:
        print(param)
        mc_simul = mcmc[param]
        mean = np.mean(mc_simul)
        sd = np.std(mc_simul)
        
        test_stat = (mean-params[param])/sd * np.sqrt(j)
        print(mean)
        print(sd)
        print(test_stat)
    print('\n')
```
