```python
def credible_interval(x, alpha):
    n = len(x)
    sided = int(alpha/2 * n)
    x = sorted(list(x))
    
    lower = x[sided]
    upper = x[n-sided]
    return lower,upper

credible_interval(range(11230), 0.05)
```

```python
credible_interval(mcmc.params['beta1'], 0.05)
```

```python
params_name = ['beta0', 'beta1', 'mu_x', 's2_x', 's2_ep', 's2_v', 'x']
print('====================== 95% credible interval for each parameter ======================')
for n in params_name[:-1]:
    lower, upper = credible_interval(mcmc.params[n], 0.05)
    print(f'{n}:  {lower},  {upper}')
```
